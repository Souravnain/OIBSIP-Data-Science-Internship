```python
df = pd.read_csv("Unemployment_Rate_upto_11_2020.csv", encoding='utf-8')
```

```python
df = df.rename(columns={
    'Region': 'State',
    'Date': 'Date',
    'Estimated Unemployment Rate (%)': 'Unemployment_Rate',
    'Estimated Employed': 'Employed',
    'Estimated Labour Participation Rate (%)': 'Labour_Participation_Rate',
    'Area': 'Area'
})
```

```python
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True, errors='coerce')

# STEP 9: Check for missing values (optional but good)
print("\nMissing values:")
print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt

# Pehle column names clean karo
df.columns = df.columns.str.strip()

# Fir column names rename karo (make sure names exactly match)
df = df.rename(columns={
    'Region': 'State',
    'Date': 'Date',
    'Estimated Unemployment Rate (%)': 'Unemployment_Rate',
    'Estimated Employed': 'Employed',
    'Estimated Labour Participation Rate (%)': 'Labour_Participation_Rate',
    'Area': 'Area'
})

# Date format thik karo
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)

# Plotting
plt.figure(figsize=(12, 6))
for state in df['State'].unique():
    state_data = df[df['State'] == state]
    plt.plot(state_data['Date'], state_data['Unemployment_Rate'], label=state)

plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.title("Unemployment Rate by State Over Time")
plt.legend(loc='upper right', bbox_to_anchor=(1.25, 1), fontsize='small')
plt.grid(True)
plt.tight_layout()
plt.show()
```
